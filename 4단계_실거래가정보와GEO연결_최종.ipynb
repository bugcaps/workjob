## geo를 기준으로 동일 주소에 있는 아파트에 대해서 좌표정보를 갱신한다.

```python
newdata =geoadds.filter(items=['주소',	'locations',	'lon',	'lat'])
```

```python
apt
```

```python
import numpy as np
def find_locaction(address):

  temp =geoadds[geoadds.주소.str.contains(address)]
  print('['+address+']')
  print(temp )
  rst = list(temp['locations'])
  if (len(rst) == 0):
     return np.nan
  return rst[0]

def find_lon(address):

  temp =geoadds[geoadds.주소.str.contains(address)]

  rst = list(temp['lon'])
  if (len(rst) == 0):
     return np.nan
  return rst[0]  

def find_lat(address):

  temp =geoadds[geoadds.주소.str.contains(address)]

  rst = list(temp['lat'])
  if (len(rst) == 0):
     return np.nan
  return rst[0]  
```

```python
aptgeo =pd.merge(left = apt , right = newdata, how = "outer", on = "주소")
aptgeo.lat 
```
